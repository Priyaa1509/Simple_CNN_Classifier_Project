```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(breast_cancer_dataset.data , columns =breast_cancer_dataset.feature_names)
```

```python
#adding target column
df['label'] = breast_cancer_dataset.target
```

```python
X = df.drop(columns = 'label',axis = 1)
X
```

```python
y = df['label']
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2 , random_state = 5)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Stab=ndardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) #fit data only to training data
X_test_std = scaler.transform(X_test)

X_train_std
```

```python
tf.random.set_seed(3) #To obtain consistency in the answer
from tensorflow import keras #Easy coding
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Flatten

model = Sequential([

    # Input Layer: Specifies the shape of the input data (e.g., 28 features)
    # It ensures proper model initialization and avoids warnings
    Input(shape=(30,)),  # Replace 28 with the actual number of features in your dataset

    # First Hidden Layer: Fully connected layer with 20 neurons
    # Activation: ReLU (Rectified Linear Unit) helps introduce non-linearity
    Dense(20, activation='relu'),

    # Output Layer: Fully connected layer with 2 neurons (for binary classification)
    # Activation: Sigmoid is used for binary or multi-label classification
    Dense(2, activation='sigmoid')
])
```

```python
#compiling neural network
model.compile(optimizer = 'adam' , #adaptive moment estimation --update weight to minimize loss function during training  -- like a smart driver : Remembers past directions , checks condition and adjust accordingly
              loss = 'sparse_categorical_crossentropy', #For integers from 1 to n ---multi-class classification
              metrics = ['accuracy'])
```

```python
#training
history = model.fit(X_train_std,y_train,validation_split = 0.1 , epochs = 10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.ylabel('Epoch')
plt.legend(['train','val data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.ylabel('Epoch')
plt.legend(['train','val data'],loc = 'lower right')
plt.figure(figsize=(5,10))
```

```python
loss,accuracy = model.evaluate(X_test_std , y_test)
```
